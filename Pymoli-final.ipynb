```python
import pandas as pd
purchase_data = "Resources/purchase_data.csv"
purchase_data_df = pd.read_csv(purchase_data)
purchase_data_df.head()
```

## Player Count

* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
      
unique_items = len(purchase_data_df["Item ID"].unique())        

average_price = purchase_data_df["Price"].mean()

sum_price = purchase_data_df["Price"].sum()

number_of_purchases = len(purchase_data_df["Purchase ID"])
# number_of_purchases = purchase_data_df["Purchase ID"].count()
                          
#sum_price_df = purchase_data_df.iloc[:, 6:].sum(axis=1)
#total_revenue = sum(sum_price_df['Total Reviews'])  
# sum_price_df= purchase_data_df["Price"].sum()

                          
                          
purchase_analysis_df = pd.DataFrame({"Number of Unique Items":[unique_items],
                                     "Average Price":[average_price],
                                     "Number of Purchases":[number_of_purchases],
                                    "Total Revenue":[sum_price],
                                    })
purchase_analysis_df["Average Price"] = purchase_analysis_df["Average Price"].map("${:.2f}".format)
purchase_analysis_df["Total Revenue"] = purchase_analysis_df["Total Revenue"].map("${:.2f}".format)
purchase_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 999]
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

gender_df["Age Range"] = pd.cut(gender_df["Age"], bins, labels=group_labels)
gender_df

age_count = gender_df["Age Range"].value_counts()
age_percent  = age_count/total_players * 100
age_count
age_percent 

age_demographics_df = pd.DataFrame({"Total Count":age_count,
                                       "Percentage of Players":age_percent 
                                   })
age_demographics_df = age_demographics_df.sort_index()
age_demographics_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data_df["Age Range"] = pd.cut(purchase_data_df["Age"], bins, labels=group_labels)
purchase_data_df

pur_analysis = purchase_data_df.groupby(["Age Range"]).count()["Purchase ID"]
pur_analysis

pur_price = purchase_data_df.groupby(["Age Range"]).mean()["Price"]

total_price = purchase_data_df.groupby(["Age Range"]).sum()["Price"]
total_price

pur_per_person = purchase_data_df.groupby(["Age Range"]).sum()["Price"]
pur_per_person

pur_analysis_final_df = pd.DataFrame({"Purchase Count":pur_analysis,
                                   "Avg. Purchase Price":pur_price,
                                   "Total Purchase Value":total_price,
                                   "Avg. Purchase Total Per Person":pur_per_person
                                  })
pur_analysis_final_df["Avg. Purchase Price"] = pur_analysis_final_df["Avg. Purchase Price"].map("${:.2f}".format)
pur_analysis_final_df["Total Purchase Value"] = pur_analysis_final_df["Total Purchase Value"].map("${:.2f}".format)
pur_analysis_final_df["Avg. Purchase Total Per Person"] = pur_analysis_final_df["Avg. Purchase Total Per Person"].map("${:.2f}".format)
pur_analysis_final_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python

top_spenders = purchase_data_df.groupby(["SN"]).count()["Purchase ID"]
avg_pur_price = purchase_data_df.groupby(["SN"]).mean()["Price"]
total_top_pur = purchase_data_df.groupby(["SN"]).sum()["Price"]

top_spenders_final_df = pd.DataFrame({"Purchase Count":top_spenders,
                                   "Avg. Purchase Price":avg_pur_price,
                                   "Total Purchase Value":total_top_pur
                                  })
top_spenders_final_df.sort_values(by=['Purchase Count'], inplace=True, ascending=False)
top_spenders_final_df["Avg. Purchase Price"] = top_spenders_final_df["Avg. Purchase Price"].map("${:.2f}".format)
top_spenders_final_df["Total Purchase Value"] = top_spenders_final_df["Total Purchase Value"].map("${:.2f}".format)

top_spenders_final_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_df = purchase_data_df[["Item ID", "Item Name","Price"]]

purchase_count = purchase_data_df.groupby(["Item ID", "Item Name"]).count()["Purchase ID"]
item_price = purchase_data_df.groupby(["Item ID", "Item Name"]).mean()["Price"]
total_pur_value = purchase_data_df.groupby(["Item ID", "Item Name"]).sum()["Price"]

popular_df = pd.DataFrame({"Purchase Count":purchase_count,
                                   "Item Price":item_price,
                                   "Total Purchase Value":total_pur_value
                          })
popular_df.sort_values(by=['Purchase Count'], inplace=True, ascending=False)
popular_df["Item Price"] = popular_df["Item Price"].map("${:.2f}".format)
popular_df["Total Purchase Value"] = popular_df["Total Purchase Value"].map("${:.2f}".format)
popular_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting

```python
popular_df = pd.DataFrame({"Purchase Count":purchase_count,
                                   "Item Price":item_price,
                                   "Total Purchase Value":total_pur_value
                          })
popular_df.sort_values(by=['Total Purchase Value'], inplace=True, ascending=False)
popular_df["Item Price"] = popular_df["Item Price"].map("${:.2f}".format)
popular_df["Total Purchase Value"] = popular_df["Total Purchase Value"].map("${:.2f}".format)
popular_df
```
